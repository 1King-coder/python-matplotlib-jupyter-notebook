```python
from modules.Theorem import Point, Straight
import matplotlib.pyplot as plt
"""
Here I use the theorem to build a circunference.
"""

def my_plotter(ax, data1, data2, param_dict={ 'marker':'o' }):
    return ax.plot(data1, data2, **param_dict)

fig, ax = plt.subplots(1, 1)

center = Point((0, 0))

def circunference(center: Point, radius: float, rel_points:list = []):
    for angle in range(0, 180):
        rel_points.append(Straight.rel_points(center, radius, angle))

        

    for point in rel_points:
        my_plotter(ax, (center.x, point[0].x), (center.y, point[0].y))
        my_plotter(ax, (center.x, point[1].x), (center.y, point[1].y))
        

circunference(center, 2)
```
